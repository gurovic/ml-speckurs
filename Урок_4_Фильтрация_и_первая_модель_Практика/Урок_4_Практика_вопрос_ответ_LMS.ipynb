**Вопрос.** Задание 1 [2 балла]. Сколько девочек сдали ЕГЭ (пол == 'Ж' и сдал == 'да')?

**Вопрос.** Задание 2 [2 балла]. Сколько мальчиков из Москвы (пол == 'М' и город == 'Москва') — через .query()?

**Ответ:** 61


**Вопрос.** Задание 3 [2 балла]. Средний балл ЕГЭ у тех, кто с репетитором (округлить до одного знака)?

**Ответ:** 68.9


**Вопрос.** Задание 4 [2 балла]. Сколько учеников готовились > 10 часов в неделю И сдали на 80 баллов и выше?

**Ответ:** 172


**Вопрос.** Задание 5 [2 балла]. Сколько учеников НЕ из Москвы и НЕ из Санкт-Петербурга (через .isin и ~)?

**Ответ:** 746


**Вопрос.** Задание 6 [2 балла]. Сколько лицеистов с мотивацией >= 8 (через .query())?

**Ответ:** 53


**Вопрос.** Задание 7 [3 балла]. Убрав пропуски в 'часов_сна', средний балл ЕГЭ у спящих >= 8 ч (округлить до одного знака)?

**Ответ:** 72.7


**Вопрос.** Задание 8 [3 балла]. На сколько баллов средний балл ЕГЭ с репетитором выше, чем без (округлить до одного знака)?

**Ответ:** 3.4

**Вывод:** с репетитором балл ~3.4 выше — эффект есть, но небольшой.


**Вопрос.** Задание 9 [3 балла]. Сколько человек попадают одновременно в группы «трудяги» и «столичные»?

**Ответ:** 35


**Вопрос.** Задание 10 [3 балла]. Чему равна разница средних баллов ЕГЭ (лицей − обычная), округлённая до одного знака?

**Ответ:** −0.7

**Вывод:** разница отрицательная (−0.7) — лицеисты в среднем сдают чуть хуже обычной школы.


**Вопрос.** Задание 11 [2 балла]. На сколько баллов (целая часть) высокомотивированные (мотивация >= 8) опережают низкомотивированных (мотивация <= 3)?

**Ответ:** 28


**Вопрос.** Задание 12 [4 балла]. Какой балл ЕГЭ предскажет LinearRegression на признаках 'средний_балл_года' и 'пропусков_уроков' для (3.5, 5), округлённый до целого?
