# Урок 2. Практика — вопросы и ответы

Краткая справка по 12 заданиям практики: формулировка, балл, эталонный ответ.


**Вопрос.** Задание 1 [1 балл]. Сколько всего строк и столбцов в таблице `df`?

**Ответ:** 1000 строк, 16 столбцов


**Вопрос.** Задание 2 [2 балла]. Как зовут ученика в строке с меткой 500 (используйте `.loc`)?

**Ответ:** Михаил


**Вопрос.** Задание 3 [2 балла]. Какой максимальный и какой минимальный балл ЕГЭ в таблице?

**Ответ:** максимум 100, минимум 20.
**Вывод:** разрыв в 80 баллов показывает, что подготовка учеников очень разная — от совсем слабой до отличной.


**Вопрос.** Задание 4 [2 балла]. Чему равен средний балл ЕГЭ по всем ученикам (округлить до одного знака)?

**Ответ:** 67.1


**Вопрос.** Задание 5 [2 балла]. Какой любимый предмет у ученика в строке с меткой 100?

**Ответ:** Физика


**Вопрос.** Задание 6 [3 балла]. Из какого города первый ученик таблицы (строка 0), если достать значение через `.iloc`?

**Ответ:** Нижний Новгород


**Вопрос.** Задание 7 [2 балла]. Сколько часов в неделю готовился ученик с самой долгой подготовкой?

**Ответ:** 26.8


**Вопрос.** Задание 8 [3 балла]. Возьмите последние 100 учеников таблицы. Чему равен их средний балл ЕГЭ (округлить до одного знака)?

**Ответ:** 63.4


**Вопрос.** Задание 9 [3 балла]. Каков максимальный возраст среди первых 250 учеников?

**Ответ:** 18


**Вопрос.** Задание 10 [3 балла]. Возьмите учеников с позиции 300 по 399. Чему равно их среднее число часов сна (округлить до одного знака)?

**Ответ:** 7.1


**Вопрос.** Задание 11 [3 балла]. Возьмите учеников с позиции 100 по 249 (150 человек). Чему равно их среднее количество пробников (округлить до одного знака)?

**Ответ:** 6.2


**Вопрос.** Задание 12 [4 балла]. Возьмите строки с меткой 500 по 600 включительно (через `.loc`). Чему равен их средний балл за год (округлить до двух знаков)?

**Ответ:** 2.98
**Вывод:** в срез `.loc[500:600]` попадает 101 строка (правая граница включается).
