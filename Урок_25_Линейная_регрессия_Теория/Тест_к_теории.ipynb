## 1. Прямая по двум точкам

**Тип:** открытый ответ

## 4. Baseline

**Тип:** выбор ответа

**Задание.** Зачем в задаче регрессии считать baseline, который всегда предсказывает среднее значение цели на train?

**Варианты ответа:**

A. Чтобы получить точку сравнения для более сложных моделей.  
B. Чтобы заменить train на test.

## 5. Интерпретация коэффициентов

**Тип:** сопоставление

**Задание.** Модель цены квартиры использует признаки `площадь_м2`, `до_центра_км`, `этаж`. Коэффициенты: `[0.18, -0.25, 0.03]`. Сопоставьте коэффициент с интерпретацией при прочих равных.

**Варианты для сопоставления:**

A. `0.18` при `площадь_м2`  
B. `-0.25` при `до_центра_км`  
C. `0.03` при `этаж`

1. при увеличении расстояния до центра прогноз цены уменьшается;  
2. при увеличении площади прогноз цены увеличивается;  
3. при увеличении этажа прогноз цены немного увеличивается.

**Правильный ответ:** A–2, B–1, C–3.
