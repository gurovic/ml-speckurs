# Занятие 16. Практика — вопрос → ответ

Справка для проверяющих. Датасет `data/orders.csv` — 2000 заказов службы доставки еды.
Вопрос заказчика: клиенты не возвращаются, что мешает зарабатывать. Всего 10 заданий, 30 баллов.

**Вопрос.** Задание 1 [2 балла]. Загрузить `orders.csv`, вывести размер и `.describe()` по `delivery_time`.

**Ответ:** Размер (2000, 8). min 0.5, max 519.3, среднее 43.0, медиана 35.5, std 45.5.

**Вывод:** Доставка за 0.5 и за 519 минут невозможна — в данных аномалии. Среднее заметно выше медианы: крайние значения тянут его вверх.

**Вопрос.** Задание 2 [3 балла]. Найти выбросы `delivery_time` по правилу 1.5 · IQR: Q1, Q3, IQR, границы, количество.

**Ответ:** Q1 = 29.8, Q3 = 43.1, IQR = 13.3. Границы нормы: 9.8…63.1. Выбросов — 211.

**Вопрос.** Задание 3 [3 балла]. Найти выбросы по Z-score (|z| > 3), сравнить с IQR, объяснить расхождение.

**Ответ:** 32 выброса, минимальное время среди них 280.6 мин. std = 45.5 при медиане 35.5.

**Вывод:** Z-score нашёл 32 заказа против 211 по IQR. std раздут самими «зависшими» заказами и почти равен медиане, поэтому порог среднее + 3·std уехал вправо: выбросы портят меру, которой их ищут.

**Вопрос.** Задание 4 [3 балла]. Разобраться в природе аномалий: заказы < 5 мин и > 180 мин, какие курьеры их везли.

**Ответ:** 45 заказов быстрее 5 минут; 32 заказа дольше 180 минут, из них 28 — вело, 4 — авто.

**Вывод:** Это сбои приложения, а не реальная доставка: часть курьеров нажала «доставлено» сразу, а 28 из 32 «зависших» заказов — у велокурьеров, то есть баг в их версии приложения. Строки надо удалить.

**Вопрос.** Задание 5 [2 балла]. Применить правило 1.5 · IQR к `order_sum`: граница, число заказов за ней, сколько из них дороже 8000 руб.

**Ответ:** Верхняя граница 1682 руб, за ней 97 заказов, из них 55 — корпоративные (дороже 8000 руб).

**Вывод:** Удалять нельзя. Это настоящие редкие значения — корпоративные и крупные семейные заказы, самые прибыльные клиенты. Сумма заказа скошена вправо, и IQR штампует «выбросы» из честных данных.

**Вопрос.** Задание 6 [3 балла]. Собрать таблицу `clean` (delivery_time от 5 до 180 включительно), сравнить среднее до и после.

**Ответ:** Было 2000 строк, стало 1923. Среднее время доставки: 43.0 → 38.4 минуты.

**Вывод:** Удалено 77 строк (менее 4% данных), а среднее упало почти на 5 минут: отчёт по сырым данным завышал время доставки.

**Вопрос.** Задание 7 [3 балла]. Построить boxplot и violinplot времени доставки по районам, найти виновный район.

**Ответ:** Медианы: Заречный 57.6, Северный 35.6, Приморский 33.4, Центр 31.9.

**Вывод:** Выбивается Заречный: медиана 57.6 против 32–36. На violin у него видны два пика — часть адресов обслуживается быстро, часть (за мостом) уходит вправо. Boxplot этого не показывает.

**Вопрос.** Задание 8 [4 балла]. t-тест пеший против вело на сырых и на чистых данных плюс Манна–Уитни на сырых.

**Ответ:** Сырые (t-тест): 46.2 против 49.0, p = 0.337. Чистые (t-тест): 46.8 против 37.4, p ≈ 4.9e-38. Сырые (Манна–Уитни): медианы 42.55 и 35.3, p ≈ 1.2e-31.

**Вывод:** На сырых данных t-тест соврал: вело выглядят «медленнее» и разница незначима — 28 зависших заказов велокурьеров раздули их среднее. После чистки вывод переворачивается: пешие медленнее на 9 минут. Манна–Уитни дал верный ответ даже на грязных данных, потому что работает с рангами.

**Вопрос.** Задание 9 [4 балла]. χ²: влияет ли долгая доставка (> 45 мин) на повторный заказ.

**Ответ:** Доля повторных: 0.448 при быстрой доставке и 0.221 при долгой. chi2 = 66.9, p ≈ 2.9e-16.

**Вывод:** H₀ — «скорость доставки не влияет на повторный заказ». Доля вернувшихся падает вдвое, p много меньше 0.05, H₀ отвергаем: долгое ожидание убивает повторные заказы.

**Вопрос.** Задание 10 [3 балла]. χ²: влияет ли промокод на повторный заказ. Финальный ответ заказчику.

**Ответ:** Доли 0.398 (промо) и 0.402 (без промо). chi2 = 0.01, p = 0.908. H₀ не отвергаем. Повторные заказы по районам: Заречный 0.251 против 0.428–0.440 у остальных.

**Вывод:** Промокод не влияет на возврат клиента — деньги потрачены впустую. Ответ заказчику: клиенты не возвращаются из-за долгой доставки (при ожидании дольше 45 минут доля повторных падает вдвое). Источник задержек — район Заречный и пешие курьеры. Нужно перевести Заречный на авто- и велодоставку вместо промокодов и починить приложение велокурьеров.
