# Занятие 33. Мини-задачки: bagging и случайный лес

Короткая проверка понимания теории из `Урок_33_Ансамбли_Bagging_Случайный_лес.ipynb`.

Все задания приведены к формату для LMS. У каждого задания явно указан тип, варианты ответа там, где они нужны, и правильный ответ.


## 1. Идея bagging

**Тип:** выбор ответа

**Задание.** Почему усреднение нескольких разных деревьев часто даёт более устойчивый прогноз, чем одно дерево?

**Варианты ответа:**

A. Усреднение снижает разброс, потому что ошибки разных деревьев частично компенсируют друг друга.  
B. Усреднение делает все деревья одинаковыми.  
C. Bagging запрещает ошибаться каждому отдельному дереву.  
D. Bagging всегда удаляет все шумные признаки.

**Правильный ответ:** A.


## 2. Bootstrap

**Тип:** выбор ответа

**Задание.** Bootstrap-выборка строится с возвращением. Может ли один и тот же объект попасть в bootstrap-выборку несколько раз, а другой объект не попасть совсем?

**Варианты ответа:**

A. Да, один объект может попасть несколько раз, а другой может не попасть совсем.  
B. Нет, каждый объект попадает ровно один раз.  
C. Один объект может попасть несколько раз, но все остальные тоже обязаны попасть.  
D. Bootstrap не связан с выборкой объектов.

**Правильный ответ:** A.
