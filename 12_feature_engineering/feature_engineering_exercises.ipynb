# Занятие 23. Мини-задачки: конструирование признаков

Короткая проверка понимания теории из `feature_engineering_theory.ipynb`.

Формат для LMS: текстовые задания без кода. Ответы должны быть короткими, но в задачах с выбором признака нужно добавить одно объясняющее предложение.


## 1. Безопасный признак

Мы предсказываем цену квартиры. Какие признаки безопасны, а какие дают утечку?

1. `площадь / комнаты`;
2. `цена / площадь`;
3. `расстояние до центра`;
4. `разница между реальной ценой и средней ценой района`.

Для каждого пункта напишите «можно» или «нельзя» и объясните, где используется неизвестная в момент прогноза цель.


## 2. Нормировка на размер или длительность

Два клиента сделали по 120 покупок. За первым наблюдали 30 дней, за вторым — 120 дней.

1. Почему число `120 покупок` не позволяет честно сравнить клиентов?
2. Какой признак лучше построить, чтобы учесть разную длительность наблюдения?


## 3. Логарифм для длинного хвоста

Признак `просмотры_за_месяц` принимает значения 0, 100, 10 000.

Почему для такого признака иногда используют `log1p`, а не оставляют исходные числа? Ответьте без вычислений: какую проблему это смягчает?


## 4. Признак-взаимодействие

Гипотеза: балкон сильнее влияет на цену большой квартиры, чем маленькой.

Какой новый признак можно построить из `площадь` и `есть_балкон`, чтобы модель могла заметить такую зависимость?


## 5. Признаки из даты

Момент прогноза — 1 июня. Мы предсказываем, купит ли клиент товар в июне.

Какие признаки можно использовать?

1. месяц регистрации клиента;
2. число покупок клиента до 1 июня;
3. число покупок клиента за весь июнь.

Для каждого пункта напишите «можно» или «нельзя».


## 6. One-hot или ordinal

Для каких признаков лучше подходит ordinal encoding, а для каких — one-hot?

1. состояние квартиры: требует ремонта / среднее / хорошее;
2. район города: центр / север / юг / запад;
3. размер футболки: S / M / L / XL.

Коротко объясните, где есть естественный порядок, а где его нет.


## 7. Пропуск как сигнал

В признаке `до_метро_минут` часть значений пропущена. Пропуск часто означает, что метро рядом нет или его не указали в объявлении.

Почему кроме заполнения пропусков медианой может быть полезен отдельный признак `метро_не_указано`?


## 8. Неизвестная категория

Модель обучалась на районах `центр`, `север`, `юг`. В test появился район `восток`.

Что должно произойти при one-hot encoding: модель должна упасть с ошибкой или спокойно обработать неизвестный район? Какой смысл у параметра `handle_unknown="ignore"`?


## 9. Агрегаты без утечки

Мы строим признак «средняя цена прошлых продаж в районе». Момент прогноза — 1 июня.

Какие продажи можно включить в расчёт: до 1 июня, после 1 июня или все продажи сразу? Объясните почему.


## 10. Проверка полезности признака

Есть два варианта кодирования признака `район`: one-hot и target encoding, посчитанный только на train.

Как честно сравнить эти варианты: по train, validation или test? Что должно быть одинаковым при сравнении?
