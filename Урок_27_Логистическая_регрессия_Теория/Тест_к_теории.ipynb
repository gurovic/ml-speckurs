## 12. * Калибровка вероятностей

**Тип:** выбор ответа

**Задание.** Модель часто выдаёт вероятность около 0,8. Среди объектов с такой вероятностью класс 1 встречается только в 60% случаев. Можно ли сказать, что вероятности хорошо откалиброваны?

**Варианты ответа:**

A. Да, потому что 0,8 больше 0,5.  
B. Нет, потому что вероятность 0,8 должна примерно соответствовать 80% реальных срабатываний.  
C. Да, потому что калибровка не связана с вероятностями.  
D. Нельзя ответить, потому что не указан solver.

**Правильный ответ:** B.
