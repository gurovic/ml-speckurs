# Занятие 35. Мини-задачки: градиентный бустинг

Короткая проверка понимания теории из `35_Градиентный_бустинг.ipynb`.

Все задания приведены к формату для LMS. У каждого задания явно указан тип, варианты ответа там, где они нужны, и правильный ответ.


## 1. Последовательный ансамбль

**Тип:** сопоставление

**Задание.** Сопоставьте описание с типом ансамбля.

**Варианты для сопоставления:**

A. Модели обучаются в основном независимо.  
B. Модели обучаются последовательно, и новая модель учитывает ошибки предыдущих.

1. bagging  
2. boosting

**Правильный ответ:** A–1, B–2.


## 2. Остатки в регрессии

**Тип:** открытый ответ

**Задание.** Факт равен 15, текущий прогноз ансамбля равен 12. Какой остаток `y - ŷ` нужно исправить следующему дереву?

**Формат ответа:** одно число.

**Правильный ответ:** `3`.


## 3. Learning rate

**Тип:** открытый ответ

**Задание.** Дерево предложило исправление `+5`, learning rate равен 0,2. На сколько изменится прогноз ансамбля?

**Формат ответа:** одно число.

**Правильный ответ:** `1`.


## 4. Маленький learning rate

**Тип:** выбор ответа

**Задание.** Что обычно верно для маленького learning rate в градиентном бустинге?

**Варианты ответа:**

A. Нужно больше деревьев; обучение дольше, но шаги осторожнее.  
B. Нужно меньше деревьев; модель всегда становится хуже.  
C. Learning rate перестаёт влиять на прогноз.  
D. Все деревья становятся независимыми, как в bagging.

**Правильный ответ:** A.


## 5. Сложность базовых деревьев

**Тип:** выбор ответа

**Задание.** Что сильнее рискует переобучиться в бустинге?

**Варианты ответа:**

A. Деревья глубины 2.  
B. Деревья глубины 12.  
C. Деревья без признаков.  
D. Глубина дерева не влияет на сложность.

**Правильный ответ:** B.


## 6. Early stopping

## 8. Масштабирование

**Тип:** выбор ответа

**Задание.** Нужно ли деревьям в градиентном бустинге обязательное масштабирование признаков?

A. Нет, потому что деревья делают разбиения по порогам признаков.
