# Занятие 12. Практика: анализ формы и плотности — вопрос → ответ

Короткая справка для проверки: что построить и какой вывод ожидается. Максимум — 30 баллов.

**Вопрос.** Задание 1 [2 балла]. Постройте **гистограмму с кривой плотности** (`histplot` с `kde=True`) для зарплаты (`salary`) и посчитайте её асимметрию (`.skew()`). Опишите форму распределения словами.

**Ответ:** скос вправо, skew ~1.48

**Вывод:** Распределение зарплат сильно скошено вправо (skew ≈ 1.5): большинство получает умеренно, а немногочисленные высокие зарплаты тянут длинный правый хвост.

**Вопрос.** Задание 2 [3 балла]. **Логарифмируйте** зарплату: создайте столбец `log_salary = np.log(df['salary'])`, постройте его гистограмму с KDE и сравните асимметрию **до и после** логарифмирования. Что стало с формой?

**Ответ:** skew 1.48 → 0.23, почти нормально

**Вывод:** Логарифм превратил скошенное распределение в почти симметричное: асимметрия упала с ~1.48 до ~0.23, гистограмма стала похожа на «колокол». Так скошенные признаки приводят к нормальному виду.

**Вопрос.** Задание 3 [3 балла]. Сравните два преобразования скошенной зарплаты — **корень** (`np.sqrt`) и **логарифм** (`np.log`). Постройте оба распределения и посчитайте асимметрию каждого. Какое преобразование лучше приближает к нормальному виду?

**Ответ:** sqrt 0.84, log 0.23 → log лучше

**Вывод:** Корень уменьшил скос лишь частично (skew ~0.84), а логарифм почти убрал его (skew ~0.23). Для сильно скошенных вправо данных логарифм нормализует лучше корня.

**Вопрос.** Задание 4 [3 балла]. Постройте **графики плотности** (`kdeplot`) зарплаты **по грейдам** (`hue='grade'`) и посчитайте медианную зарплату каждого грейда. Как отличаются распределения грейдов по типичному уровню и разбросу?

**Ответ:** медианы Junior 87k → Lead 411k, разброс растёт

**Вывод:** С ростом грейда медианная зарплата растёт (Junior ~87k → Lead ~411k), а распределение смещается вправо и расширяется: у Lead и разброс, и уровень заметно выше — грейд почти полностью определяет зарплату.

**Вопрос.** Задание 5 [3 балла]. Исследуйте форму признака **`experience`** (опыт): постройте гистограмму с KDE и посчитайте асимметрию. Это распределение скошенное или симметричное? В какую сторону?

**Ответ:** скос вправо, skew ~0.82

**Вывод:** Опыт скошен вправо, но слабее зарплаты (skew ≈ 0.8): преобладают сотрудники с небольшим стажем, а очень опытных мало — они образуют правый хвост.

**Вопрос.** Задание 6 [4 балла]. Составьте **«карту скошенности»** данных: посчитайте асимметрию (`.skew()`) для всех числовых признаков — `salary`, `bonus`, `age`, `experience`, `performance` — и отсортируйте (можно вывести таблицей или горизонтальными столбцами). Какие признаки сильно скошены **вправо** (кандидаты на логарифм), какие почти симметричны, а какой скошен **влево**?

**Ответ:** bonus 2.06, salary 1.48 (вправо); performance −0.63 (влево)

**Вывод:** Сильнее всех скошены вправо bonus (~2.06) и salary (~1.48) — их стоит логарифмировать. Опыт и возраст скошены умеренно, а performance скошен **влево** (~−0.63): большинство оценок высокие (4–5), а низкие образуют левый хвост. Знак и величина skew сразу подсказывают, что и как преобразовывать.

**Вопрос.** Задание 7 [3 балла]. Сравните распределение зарплат для **удалёнщиков и офисных** (`kdeplot` с `hue='remote'`) и посчитайте медиану каждой группы. Есть ли между ними реальная разница?

**Ответ:** медианы ~143k≈145k — разницы нет

**Вывод:** Кривые плотности почти совпадают, медианы близки (~143k и ~145k) — заметной разницы нет. Важный вывод: не всякий срез данных даёт различие; формат работы (удалёнка/офис) на зарплату здесь не влияет.

**Вопрос.** Задание 8 [3 балла]. Исследуйте **бонусы** (`bonus`): постройте распределение с KDE, посчитайте асимметрию, затем примените логарифм (`np.log1p` — он безопасен к нулям) и снова посчитайте асимметрию. Помог ли логарифм — и не перестарался ли он?

**Ответ:** 2.06 → −1.14, логарифм перекосил влево

**Вывод:** Бонусы сильно скошены вправо (skew ~2.06). Логарифм уменьшил скос по модулю, но перестарался — асимметрия стала отрицательной (~−1.14), то есть распределение перекосило влево. Вывод: преобразование нужно подбирать и проверять по числу, а не применять вслепую.

**Вопрос.** Задание 9 [3 балла]. Посмотрите, как логарифм влияет на **связь** опыта и зарплаты. Постройте два точечных графика: `experience` vs `salary` и `experience` vs `log_salary`. На каком связь читается чётче и почему?

**Ответ:** воронка → ровная полоса

**Вывод:** На сырой зарплате точки расходятся «воронкой» — с ростом опыта увеличивается и разброс. После логарифма разброс выравнивается, и восходящая связь опыта и зарплаты читается как ровная полоса. Логарифм не только нормализует форму, но и стабилизирует разброс.

**Вопрос.** Задание 10 [3 балла]. Итоговое исследование. Найдите отдел с самой **высокой** и самой **низкой** медианной зарплатой, постройте их распределения на одном графике (`kdeplot` с `hue`) и сравните форму и типичный уровень. Сделайте вывод 2–3 предложения.

**Ответ:** Поддержка ~106k vs Инженерия ~167k

**Вывод:** Ниже всех платят в Поддержке (медиана ~106k), выше всех — в Инженерии (~167k). Обе кривые скошены вправо, но у Инженерии пик смещён правее и хвост длиннее — там и типичный уровень выше, и больше высокооплачиваемых. Отдел заметно влияет на зарплату.
