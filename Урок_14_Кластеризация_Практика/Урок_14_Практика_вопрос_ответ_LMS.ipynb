# Занятие 14. Практика: кластеризация — вопрос → ответ (для проверки)

Краткая справка по всем 10 заданиям. Датасеты: `students.csv` (навыки студентов, задания 1–6)
и `health.csv` (показатели здоровья, задания 7–10). Параметры: `ward`, `euclidean`,
`fcluster(criterion='maxclust')`, `KMeans(random_state=42, n_init=10)`.

**Вопрос.** Задание 1 [2 балла]. Загрузить `students.csv`, задать список навыков `skills`, вывести размер.

**Ответ:** `(18, 11)` — 18 студентов: имя + 10 навыков.

**Вопрос.** Задание 2 [3 балла]. Иерархическая кластеризация по S1–S10 и дендрограмма с именами.

**Ответ:** `linkage(stu[skills], method='ward', metric='euclidean')` + `dendrogram(..., labels=stu['NAME'].tolist())`.

**Вывод:** На дереве 3 крупные ветви.

**Вопрос.** Задание 3 [3 балла]. Каменистая осыпь, определить число групп.

**Ответ:** Расстояния 32.9 → 20.9 → 10.8 → 6.8; перелом на **K = 3** (детализация — 4).

**Вывод:** Резкое падение до 3 групп, дальше плоско.

**Вопрос.** Задание 4 [3 балла]. fcluster на 3 и 4 кластера, размеры групп.

**Ответ:** 3 кластера: {1:5, 2:4, 3:9}; 4 кластера: {1:5, 2:4, 3:3, 4:6}.

**Вопрос.** Задание 5 [3 балла]. Центроиды при K=4, описание групп.

**Ответ:** `stu.groupby('cl4')[skills].mean().round(1)`.

**Вывод:** Аналитики (S1–4 ~9, S5–10 ~4); универсалы (всё ~9); гуманитарии со средней аналитикой (S1–4 ~6, S5–10 ~8); гуманитарии (S1–4 ~4, S5–10 ~8.7).

**Вопрос.** Задание 6 [3 балла]. KMeans(4) + crosstab с иерархией.

**Ответ:** `pd.crosstab(stu['cl4'], stu['kmeans4'])` — каждый кластер целиком в одном столбце K-Means.

**Вывод:** Иерархия и K-Means дали одно и то же разбиение — согласуются.

**Вопрос.** Задание 7 [3 балла]. Загрузить `health.csv`, стандартизовать 6 показателей, вывести размер.

**Ответ:** `(310, 6)`; `StandardScaler().fit_transform(hlt[feat])`.

**Вопрос.** Задание 8 [3 балла]. Иерархия на стандартизованных + каменистая осыпь, число групп риска.

**Ответ:** Расстояния 42.7 → 23.2 → 13.0 → 12.1; перелом → **K = 3**.

**Вывод:** Резкое падение до 3 групп, дальше плоско.

**Вопрос.** Задание 9 [4 балла]. fcluster(3), размеры и центроиды групп, описание уровней риска.

**Ответ:** Размеры {133, 107, 70}.

**Вывод:** Низкий риск (возраст ~45, bmi ~22, норма); умеренный (~55, bmi ~27); высокий (~66, bmi ~33, глюкоза ~145, sbp ~159). Холестерил не различает группы.

**Вопрос.** Задание 10 [3 балла]. Кластеризация без стандартизации + crosstab со стандартизованной.

**Ответ:** `pd.crosstab(hlt['cluster'], hlt['cluster_raw'])` — объекты НЕ по диагонали.

**Вывод:** Разбиения расходятся: холестерин (крупный масштаб) доминирует в расстоянии; стандартизация обязательна.
