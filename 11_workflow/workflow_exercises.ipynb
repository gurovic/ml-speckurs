## 1. Тип задачи

Для каждой ситуации выберите тип задачи: классификация, регрессия или кластеризация.

1. Разделить покупателей на группы, заранее не зная названия этих групп.
2. Предсказать, купит ли клиент подписку в следующие 30 дней.
3. Оценить цену квартиры по её описанию.

Коротко объясните, по какому признаку вы отличили эти три типа задач.


## 2. Обучение с учителем и без учителя

В какой из двух ситуаций есть обучение с учителем?

1. У нас есть история клиентов и известно, кто из них ушёл.
2. У нас есть только поведение клиентов, а готовых меток нет.

Ответьте номером ситуации и объясните, что в ней играет роль правильного ответа.


## 3. Train, validation и test

Сопоставьте части данных с их ролью:

- train;
- validation;
- test.

Роли:

1. обучить модель;
2. сравнить несколько вариантов модели или гиперпараметров;
3. один раз честно оценить уже выбранное решение.
