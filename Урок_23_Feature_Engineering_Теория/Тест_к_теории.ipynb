# Занятие 23. Мини-задачки: конструирование признаков

Короткая проверка понимания теории из `Урок_23_Feature_Engineering_Теория.ipynb`.

Все задания приведены к формату для LMS. У каждого задания явно указан тип, варианты ответа там, где они нужны, и правильный ответ.


## 1. Безопасный признак

**Тип:** сопоставление

**Задание.** Мы предсказываем цену квартиры. Сопоставьте признак с решением: можно использовать или нельзя использовать из-за утечки.

**Варианты для сопоставления:**

A. `площадь / комнаты`  
B. `цена / площадь`  
C. `расстояние до центра`  
D. `разница между реальной ценой и средней ценой района`

1. можно  
2. нельзя

**Правильный ответ:** A–1, B–2, C–1, D–2.


## 2. Нормировка на длительность

**Тип:** открытый ответ

**Задание.** Два клиента сделали по 120 покупок. За первым наблюдали 30 дней, за вторым — 120 дней. Какой признак лучше построить, чтобы честнее сравнить клиентов?

**Формат ответа:** название признака.

**Правильный ответ:** `покупки_в_день`.


## 3. Логарифм для длинного хвоста

**Тип:** выбор ответа

**Задание.** Признак `просмотры_за_месяц` принимает значения 0, 100, 10 000. Зачем иногда применяют `log1p` к такому признаку?

**Варианты ответа:**

A. Чтобы уменьшить влияние очень больших значений.  
B. Чтобы превратить задачу классификации в регрессию.  
C. Чтобы заменить все пропуски нулями.  
D. Чтобы удалить все маленькие значения.

**Правильный ответ:** A.


## 4. Признак-взаимодействие

**Тип:** открытый ответ

**Задание.** Гипотеза: балкон сильнее влияет на цену большой квартиры, чем маленькой. Какой новый признак можно построить из `площадь` и `есть_балкон`?

**Формат ответа:** формула признака.

**Правильный ответ:** `площадь * есть_балкон`.


## 5. Признаки из даты

**Тип:** сопоставление

**Задание.** Момент прогноза — 1 июня. Мы предсказываем, купит ли клиент товар в июне. Сопоставьте признак с решением: можно или нельзя использовать.

**Варианты для сопоставления:**

A. месяц регистрации клиента

## 6. One-hot или ordinal

**Тип:** сопоставление

**Задание.** Сопоставьте признак с подходящим способом кодирования.

**Варианты для сопоставления:**

A. состояние квартиры: требует ремонта / среднее / хорошее

## 7. Пропуск как сигнал

**Тип:** выбор ответа

**Задание.** В признаке `до_метро_минут` часть значений пропущена. Почему кроме заполнения пропусков медианой может быть полезен отдельный признак `метро_не_указано`?

**Варианты ответа:**

A. Потому что сам факт пропуска может нести информацию об объекте.  
B. Потому что индикатор пропуска всегда заменяет все остальные признаки.  
C. Потому что деревья решений не умеют работать с числами.  
D. Потому что медиану нельзя считать на train.

**Правильный ответ:** A.

## 10. Проверка полезности признака

**Тип:** выбор ответа

**Задание.** Есть два варианта кодирования признака `район`: one-hot и target encoding, посчитанный только на train. Как честно сравнить эти варианты?

**Варианты ответа:**

A. Сравнить качество на train, меняя одновременно модель и метрику.  
B. Сравнить качество на validation при одинаковом разбиении, модели и метрике.  
C. Выбрать вариант с большим числом столбцов.  
D. Проверить оба варианта на test много раз и выбрать лучший.

**Правильный ответ:** B.
