### Вопрос 9 (открытый, задача)

Логистическая компания сравнивает **два склада**. Замерили время доставки (в минутах)
у 10 случайных заказов с каждого склада:

```python
sever = [31, 28, 35, 30, 33, 29, 34, 32, 30, 36]   # склад «Север»
yug   = [28, 27, 33, 26, 31, 24, 32, 29, 26, 30]   # склад «Юг»
```

Среднее у «Севера» — 31.8 минуты, у «Юга» — 28.6. Данные близки к нормальным, выбросов нет.

**Гипотезы:**
- H₀: среднее время доставки на складах **одинаково**, разница в 3.2 минуты случайна;
- H₁: время доставки **различается**.

Проверьте H₀ и запишите **p-value, округлив до двух знаков** после запятой.

Подсказка — как посчитать:

```python
from scipy import stats
t_stat, p_value = stats.ttest_ind(sever, yug)
print(round(p_value, 2))
```

### Вопрос 10 (открытый, задача)

Магазин тестирует **две формы оплаты**. Каждую показали 500 покупателям и посчитали, сколько
из них довели заказ до конца:

| Форма | Не купили | Купили | Конверсия |
|---|---|---|---|
| Старая | 430 | 70 | 14% |
| Новая | 410 | 90 | 18% |

Конверсия новой формы выше — 18% против 14%. Но сравниваем мы **доли**, поэтому нужен **χ²**.

**Гипотезы:**
- H₀: форма оплаты **не влияет** на конверсию, разница в 4 процентных пункта случайна;
- H₁: конверсия у форм **разная**.

Проверьте H₀ и запишите **p-value, округлив до двух знаков** после запятой.

Подсказка — как посчитать:

```python
from scipy import stats
table = [[430, 70],    # старая форма: не купили / купили
         [410, 90]]    # новая форма
chi2, p_value, dof, expected = stats.chi2_contingency(table)
print(round(p_value, 2))
```

Обратите внимание на итог: при α = 0.05 нулевую гипотезу здесь **не отвергают**, хотя разница
конверсий видна глазом. Это тот самый случай «улик недостаточно» из урока.
