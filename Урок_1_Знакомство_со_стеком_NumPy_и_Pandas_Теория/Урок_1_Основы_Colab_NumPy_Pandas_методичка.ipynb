## Часть 1. Как устроен ноутбук (Colab)

Ноутбук состоит из **ячеек**. Есть два вида:
- **ячейка кода** — в ней пишут Python скрипты и запускают;
- **ячейка текста (Markdown)** — заголовки и пояснения.

**Как запустить ячейку кода:** щёлкнуть по ней и нажать **Shift + Enter**
(или кнопку «запустить» слева). Результат появится **прямо под ячейкой**.

Запустите ячейку ниже.


````python
print("Мой первый код в Jupyter Notebook")
2 + 2
````

**Обратите внимание на две вещи:**
1. `print(...)` вывел текст.
2. Результат последней строки (`2 + 2`) показался сам, без `print`.

Слева от ячейки появилось число в скобках, например `[1]` — это **номер запуска**.
Пока ячейка выполняется, там стоит звёздочка `[*]`.

### Диалоговый режим: переменные живут между ячейками

Ноутбук помнит, что вы делали раньше. Переменная, созданная в одной ячейке,
доступна в следующих. Запустите обе ячейки по очереди.


````python
student_name = 'Аня'
score = 87
````

````python
# эта ячейка использует переменные из предыдущей
print(student_name, 'набрала', score, 'баллов')
````

### Структура ноутбука

Хороший ноутбук читается как документ:
- **заголовки** (Markdown с `#`) делят работу на разделы;
- **текст** объясняет, что и зачем делаем;
- **код** выполняет действие;
- **результат** сразу под кодом.

Так работу легко понять и себе, и другим. Дальше мы будем придерживаться этой структуры.

### Как делать красивую структуру: разметка Markdown

Текстовые ячейки пишут на **Markdown** — это простой способ оформления. Несколько символов
превращают обычный текст в заголовки, списки и выделения. Вот главные приёмы:

| Что написать | Что получится |
|---|---|
| `# Заголовок` | большой заголовок раздела |
| `## Подзаголовок` | заголовок поменьше |
| `### Ещё меньше` | подзаголовок третьего уровня |
| `**жирный**` | **жирный** текст |
| `*курсив*` | *курсив* |
| `` `код` `` | `моноширинный` фрагмент кода |
| `- пункт` | маркированный список |
| `1. пункт` | нумерованный список |
| `> цитата` | выделенная врезка-заметка |

> **Совет:** чтобы отредактировать текстовую ячейку — щёлкните по ней **дважды**.
> Чтобы «применить» разметку и увидеть красивый результат — нажмите **Shift + Enter**,
> как и с кодом.

**Пример живой разметки.** Эта ячейка написана на Markdown. Дважды щёлкните по ней,
чтобы увидеть «исходник», а потом нажмите Shift + Enter, чтобы вернуть красивый вид.

### Как выглядит аккуратный ноутбук

Хороший ноутбук читается сверху вниз, как рассказ. Обычно повторяется один и тот же ритм:

1. **Заголовок** раздела (`##`) — о чём этот кусок работы.
2. **Текст-пояснение** — что и зачем мы сейчас сделаем.
3. **Ячейка кода** — само действие.
4. **Короткий вывод** под кодом — что получилось и что это значит.

Ниже — маленький, но полностью «оформленный» пример такого блока. Обратите внимание,
как текст, код и вывод идут друг за другом.

## Пример: считаем средний балл класса

Возьмём баллы нескольких учеников и посчитаем **средний балл**.
Это простая задача, но она показывает правильный порядок: сначала данные, потом расчёт, потом вывод.


````python
class_scores = [78, 85, 90, 66, 74]     # баллы учеников
average = sum(class_scores) / len(class_scores)
print('Средний балл класса:', average)
````

**Вопрос:** как одной фразой описать, что мы увидели?

## Пример: сколько учеников сдали

Ещё один аккуратный блок. Посчитаем, сколько учеников набрали **75 баллов и выше**.


````python
passed_count = 0
for s in class_scores:
    if s >= 75:
        passed_count += 1

print('Сдали (75+):', passed_count, 'из', len(class_scores))
````

**Вывод:** трое из пяти набрали 75 и больше.

## Часть 2. NumPy — быстрые вычисления с числами

`NumPy` — библиотека для работы с **массивами** чисел. Сначала подключим её.
Приставку `np` пишут по традиции — так короче.


````python
import numpy as np
````

### Массив (ndarray) вместо списка

Обычный список Python и массив NumPy выглядят похоже, но массив **быстрее** и удобнее
для вычислений: в нём все числа одного типа и лежат рядом в памяти.


````python
py_list = [1, 2, 3, 4, 5]           # обычный список Python
np_array = np.array([1, 2, 3, 4, 5])  # массив NumPy

print('py_list:', py_list)
print('np_array:', np_array)
print('тип np_array:', type(np_array))
````

### Векторизация: операции без циклов

Главное преимущество: с массивом можно работать **целиком**, одной строкой,
без цикла `for`. Это называется **векторизация**.


````python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

print('a + b =', a + b)   # поэлементное сложение
print('a * 2 =', a * 2)   # каждый элемент умножен на 2
print('a ** 2 =', a ** 2) # каждый в квадрате
````

**Вопрос:** а как сделать то же самое с обычным списком Python? Что пришлось бы написать?

### Свойства массива: dtype и shape

- **`dtype`** — тип данных внутри массива (целые, дробные и т. д.);
- **`shape`** — форма: сколько строк и столбцов.


````python
m = np.array([[1, 2, 3],
              [4, 5, 6]])
print('dtype (тип данных):', m.dtype)
print('shape (форма):', m.shape)   # (2 строки, 3 столбца)
print('всего элементов:', m.size)
````

### Агрегаты без циклов: sum, mean, min, max

Массив умеет сам считать сумму, среднее, минимум и максимум.


````python
grades = np.array([90, 75, 60, 82, 95, 70])

print('сумма :', grades.sum())
print('среднее:', grades.mean())
print('минимум:', grades.min())
print('максимум:', grades.max())
````

На следующем занятии мы увидим, что это работает мгновенно даже на **миллионе** чисел —
там же сравним скорость со списком через `%timeit`.

## Часть 3. Pandas — таблицы поверх NumPy

`Pandas` добавляет удобные **таблицы**. Подключим его (приставка `pd`).


````python
import pandas as pd
````

### Series — это один столбец

`Series`  — это **один столбец данных**. У каждого значения есть свой
**индекс** — подпись строки. Можно думать про `Series` как про два связанных списка:
слева — подписи (индекс), справа — значения.

```
      Series "scores"
   ┌─────────┬─────────┐
   │ индекс  │ значение│
   ├─────────┼─────────┤
   │ Аня     │   90    │
   │ Боря    │   75    │
   │ Вика    │   60    │
   └─────────┴─────────┘
     (подписи)  (данные)
```

**Вопрос:** чем это отличается от обычного массива NumPy? Что нового появилось слева?


````python
scores = pd.Series([90, 75, 60], index=['Аня', 'Боря', 'Вика'])
scores
````

````python
# обращение по индексу — по имени
print('Балл Ани:', scores['Аня'])
print('Среднее по всем:', scores.mean())
````

### DataFrame — это таблица из нескольких Series

`DataFrame` (читается «дата-фрейм») — это **целая таблица**. Устроена она просто:
**каждый столбец таблицы — это отдельный `Series`**, а все столбцы «склеены» рядом
и делят **один общий индекс** (подписи строк).

Схематично: берём несколько столбцов-`Series` и ставим их бок о бок —

```
   Series "имя"    Series "балл"   Series "сдал"
   ┌──────┐        ┌──────┐        ┌──────┐
 0 │ Аня  │      0 │  90  │      0 │  да  │
 1 │ Боря │      1 │  75  │      1 │  да  │
 2 │ Вика │      2 │  60  │      2 │ нет  │
 3 │ Гена │      3 │  82  │      3 │  да  │
   └──────┘        └──────┘        └──────┘
       │               │               │
       └───────────────┼───────────────┘
                       ▼
              склеиваем в DataFrame
   ┌───┬──────┬──────┬──────┐
   │   │ имя  │ балл │ сдал │   ← названия столбцов (это бывшие имена Series)
   ├───┼──────┼──────┼──────┤
   │ 0 │ Аня  │  90  │  да  │
   │ 1 │ Боря │  75  │  да  │
   │ 2 │ Вика │  60  │ нет  │
   │ 3 │ Гена │  82  │  да  │
   └───┴──────┴──────┴──────┘
     ▲                        
   индекс                     
  (общие подписи строк)       
```

Поэтому когда мы пишем `pd.DataFrame({...})`, **ключи словаря** становятся названиями
столбцов, а **списки-значения** — содержимым этих столбцов (то есть `Series`).
И наоборот: если из готовой таблицы взять **один столбец** — получится обратно `Series`.


````python
df = pd.DataFrame({
    'имя':   ['Аня', 'Боря', 'Вика', 'Гена'],
    'балл':  [90, 75, 60, 82],
    'сдал':  ['да', 'да', 'нет', 'да'],
})
df
````

Проверим это на деле: соберём таблицу **из двух отдельных `Series`**.
Создадим два столбца-`Series` с одинаковым индексом (имена учеников) и склеим их
в `DataFrame` — увидим, что имена `Series` стали названиями столбцов.


````python
# два отдельных столбца-Series с общим индексом
bally = pd.Series([90, 75, 60, 82], index=['Аня', 'Боря', 'Вика', 'Гена'])
sdal  = pd.Series(['да', 'да', 'нет', 'да'], index=['Аня', 'Боря', 'Вика', 'Гена'])

# склеиваем два Series в один DataFrame (по столбцам)
tabl = pd.DataFrame({'балл': bally, 'сдал': sdal})
tabl
````

### Первый взгляд на таблицу

Несколько базовых методов, которые пригодятся всегда:
- **`.head()`** — первые строки;
- **`.shape`** — размер (строки, столбцы);
- **`.columns`** — названия столбцов;
- **`.dtypes`** — типы данных в столбцах.


````python
print('Размер таблицы:', df.shape)
print('Столбцы:', list(df.columns))
print()
df.head()
````

**Вопрос:** если из таблицы взять один столбец — какой тип у нас получится? `DataFrame` или что-то другое?


````python
# один столбец таблицы — это Series
df['балл']
````

**Обратите внимание:** взяли из таблицы один столбец `df['балл']` — и получили
`Series`. Круг замкнулся: **DataFrame состоит из Series, а каждый его столбец — это Series.**
