# Урок 8. Практика — вопрос → ответ

Краткая справка по всем заданиям (датасет `shop_orders.csv`).

**Вопрос.** Задание 1 [1 балл]. Столбцы `created_at` и `delivered_at` пока текстовые. Преобразуйте **оба** в тип даты-времени через `pd.to_datetime()` и выведите их типы.

**Ответ:** оба столбца → datetime64[ns]

**Вопрос.** Задание 2 [2 балла]. Из `created_at` извлеките два признака в новые столбцы: `час` (`.dt.hour`) и `день_недели` (`.dt.day_name()`). Выведите первые строки этих столбцов.

**Ответ:** столбцы `час` и `день_недели`

**Вопрос.** Задание 3 [2 балла]. В какой **час суток** заказов больше всего? Найдите пиковый час и число заказов в нём (`value_counts`).

**Ответ:** 20:00 — 155 заказов

**Вывод:** Пик — 20:00 (155 заказов): заказывают вечером, после работы; днём и ночью заметно меньше.

**Вопрос.** Задание 4 [3 балла]. Постройте **матрицу активности по устройствам**: `pivot_table` с устройством (`device`) по строкам, часом (`час`) по столбцам и числом заказов в клетках (`aggfunc='count'`). По матрице видно, в какие часы заказывают с разных устройств.

**Ответ:** Desktop — днём (9–18), Mobile/Tablet — вечером (18–23)

**Вывод:** По матрице видно: с Desktop заказывают в рабочие часы (9–18), а с Mobile и особенно Tablet — вечером (18–23).

**Вопрос.** Задание 5 [2 балла]. Посчитайте **средний чек** (`amount`) по сегментам клиентов (`segment`). Отсортируйте по возрастанию.

**Ответ:** New 5890 < Returning 7885 < VIP 11086

**Вывод:** Средний чек растёт от New (5890) к VIP (11086) — почти вдвое: одна покупка VIP приносит заметно больше.

**Вопрос.** Задание 6 [4 балла]. Посчитайте **средний интервал между покупками одного клиента в разрезе сегмента**. Отсортируйте по `customer_id` и `created_at`, возьмите `.groupby('customer_id')['created_at'].diff()`, переведите в дни, затем усредните по `segment`.

**Ответ:** New 124.3, Returning 75.0, VIP 33.1 дня

**Вывод:** VIP возвращаются в среднем раз в 33 дня — почти вчетверо чаще, чем New (124 дня): самый активный и лояльный сегмент.

**Вопрос.** Задание 7 [3 балла]. Соберите **продуктовую витрину** по категориям через `groupby('category').agg(...)`: `заказов` (count), `средний_чек` (mean `amount`), `выручка` (sum `amount`), `клиентов` (nunique `customer_id`). Отсортируйте по выручке по убыванию.

**Ответ:** лидер выручки — Электроника (~4.93 млн)

**Вывод:** Больше всего выручки даёт Электроника (~4.9 млн) — не за счёт объёма, а за счёт самого высокого чека: она обгоняет более массовые Одежду и Книги.

**Вопрос.** Задание 8 [3 балла]. Найдите **самый активный день недели** — в какой день заказов больше всего. На выбор: постройте `pivot_table` день недели × час и просуммируйте по строкам, **или** сгруппируйте по `день_недели` и посчитайте заказы.

**Ответ:** Sunday — 214 (с небольшим отрывом)

**Вывод:** Формально лидирует воскресенье (214), но отрыв от других дней небольшой — по дням недели активность довольно ровная.

**Вопрос.** Задание 9 [3 балла]. Посчитайте **выручку по каналам** (`channel`, сумма `amount`) и отдельно — **долю выручки от VIP** в каждом канале (выручка VIP в канале / вся выручка канала). Какой канал приводит самых ценных клиентов?

**Ответ:** выручка — Ads; VIP-доля — Ads 0.36, Referral 0.35

**Вывод:** Ads даёт и самую большую выручку (3.3 млн), и максимальную долю VIP (36%); Referral приводит почти столько же VIP (35%) при меньшем объёме — оба канала «премиальные».

**Вопрос.** Задание 10 [2 балла]. Оцените **эффективность скидок**. Создайте признак `со_скидкой` (`discount_pct > 0`) и сравните для заказов со скидкой и без: средний чек (`amount`), среднюю корзину (`quantity`) и число заказов.

**Ответ:** со скидкой чек ниже (7383 vs 8465), корзина та же (~2.05)

**Вывод:** Скидки не увеличивают корзину (по ~2.05 позиции и там, и там), но снижают средний чек (7383 против 8465) — то есть работают как прямая потеря маржи, а не драйвер объёма.

**Вопрос.** Задание 11 [3 балла]. Разбейте сутки на **4 части** и сравните их. Создайте столбец `время_суток` из `час`: ночь (0–5), утро (6–11), день (12–17), вечер (18–23). Посчитайте по каждой части число заказов и средний чек (`groupby().agg()`). Когда заказов больше всего, а когда средний чек выше?

**Ответ:** заказов max — вечер (749); чек max — ночь (~9975)

**Вывод:** Больше всего заказов вечером (749), но средний чек там НЕ самый высокий: ночью и утром заказов мало, зато чек заметно выше (~9975 и ~9306) — редкие ночные заказы крупнее вечерних.

**Вопрос.** Задание 12 [2 балла]. Аналитика. Посчитайте **выручку на одного клиента по сегментам** (сумма `amount` / число уникальных клиентов). Опираясь на это и выводы заданий 5, 6 и 9, предложите, на какой сегмент и через какой канал выгоднее тратить маркетинг. Обоснуйте в выводе.

**Ответ:** New 9033 / Returning 28295 / VIP 105794 руб/клиент; ставка на VIP через Ads/Referral

**Вывод:** Один VIP приносит ~105 800 руб против 9 000 у New — на порядок больше, и возвращается втрое чаще. Значит выгоднее вкладываться в удержание и привлечение VIP-подобных клиентов, а самый «премиальный» канал для этого — Ads (и Referral): именно они дают наибольшую долю VIP-выручки.
