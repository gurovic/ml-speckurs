# Занятие 29. Мини-задачки: переобучение и валидация

Короткая проверка понимания теории из `Урок_29_Переобучение_и_валидация.ipynb`.

Все задания приведены к формату для LMS. У каждого задания явно указан тип, варианты ответа там, где они нужны, и правильный ответ.


## 1. Роли train, validation и test

**Тип:** сопоставление

**Задание.** Сопоставьте часть данных с действием.

**Варианты для сопоставления:**

A. train  
B. validation  
C. test

1. обучить параметры модели;  
2. выбрать степень полинома или глубину дерева;  
3. один раз оценить уже выбранную модель.

**Правильный ответ:** A–1, B–2, C–3.


## 2. Диагноз по ошибкам

**Тип:** выбор ответа

**Задание.** Train MSE = 0,01, validation MSE = 0,33. На что это больше похоже?

**Варианты ответа:**

A. недообучение  
B. хорошее обобщение  
C. переобучение

**Правильный ответ:** C.


## 3. Недообучение

**Тип:** выбор ответа

**Задание.** Train MSE и validation MSE обе большие и почти одинаковые. Что это говорит о модели?

**Варианты ответа:**

A. Модель слишком простая.  
B. Модель идеально подходит.  
C. Модель точно использует test для обучения.  
D. Модель слишком хорошо запомнила train.

**Правильный ответ:** A.


## 4. Validation curve

## 5. Test нельзя открывать много раз

**Тип:** выбор ответа

**Задание.** Команда проверила 20 вариантов модели на test и выбрала лучший. Почему после этого качество на test стало слишком оптимистичным?

**Варианты ответа:**

A. Test фактически использовали для выбора модели.  
B. Test нельзя хранить в файле CSV.  
C. Любая модель на test всегда получает качество 1.  
D. Validation и test всегда должны совпадать.

**Правильный ответ:** A.

## 7. GroupKFold

**Тип:** выбор ответа

**Задание.** В данных есть несколько фотографий одного пациента. Какой тип разбиения нужен, чтобы фото одного и того же пациента не попали одновременно в train и validation?

**Варианты ответа:**

A. Обычный случайный split по фотографиям.  
B. GroupKFold или group split по пациенту.  
C. Сортировка фотографий по имени файла.  
D. Обучение без validation.

**Правильный ответ:** B.


## 8. Time split

**Тип:** выбор ответа

**Задание.** Мы прогнозируем продажи на будущие недели. Почему нельзя перемешать прошлые и будущие недели случайным образом перед разбиением?

**Варианты ответа:**

A. Будущие данные могут попасть в train и дать утечку.  
B. Случайное перемешивание всегда удаляет все признаки.  
C. Продажи нельзя прогнозировать моделями.  
D. Time split нужен только для картинок.

**Правильный ответ:** A.
